### 逻辑回归解决二分类问题--计算事件发生的概率

Sigmoid函数 值域在(0,1)之间

```python
import numpy as np
import matplotlib.pyplot as plt
```

sigmoid函数

```python
def sigmoid(t):
    return 1 / (1 + np.exp(-t))
```

在-10到10之间取均匀取500个点的函数

```python
x = np.linspace(-10, 10, 500)
y = sigmoid(x)

plt.plot(x, y)
```
